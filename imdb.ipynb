#### Calculate the average star rating for each genre, but only include genres with at least 10 movies


**You only need to choose one of the options below.** However, solving for multiple options would count as "extra credit" that could e.g. raise your score from a 2 ("meets expectations") to a 3 ("exceeds expectations").

#### Option 1: manually create a list of relevant genres, then filter using that list

```python
# Answer:\
movies['genre'].value_counts()
movies[movies['genre'].isin(('Drama', 'Comedy', 'Action', 'Crime', 'Biography', 'Adventure', 'Animation','Horror', 'Mystery' ))].groupby('genre')['star_rating'].mean()
```

#### Option 2: automatically create a list of relevant genres by saving the value_counts and then filtering

```python
# Answer:
genres = movies['genre'].value_counts()[movies['genre'].value_counts() > 10].index
movies[movies['genre'].isin(genres)].groupby('genre')['star_rating'].mean()
```

#### Option 3: calculate the average star rating for all genres, then filter using a boolean Series

```python
# Answer:
movies.groupby('genre').star_rating.mean()[movies.genre.value_counts() >= 10]
```

#### Option 4: aggregate by count and mean, then filter using the count

```python
ratings = movies.groupby('genre').star_rating.agg(['count', 'mean'])
ratings[ratings['count']>10]
```

## Bonus

#### Figure out something "interesting" using the actors data!

```python
import ast
movies.head()
actors = movies['actors_list']
d=[]
actor_dict = {}
for i in range(0,len(actors)):
    tmp = actors[i].replace('u', '')
    tmp = ast.literal_eval(tmp)
    d.append(tmp)
    for j in tmp:
        if j in actor_dict.keys():
            actor_dict[j] += 1
        else: 
            actor_dict[j] = 1
        
```

```python
movies['actors'] = d
movies.head()
```

```python
df_actors = pd.DataFrame(actor_dict.items(), columns = ['actor', 'appearances'])
df_actors.describe()

#df_actors.dtypes()
```

```python
df_actors.hist()
```

```python
df_actors[(df_actors.loc[:,'appearances']>7)].sort_values('appearances', ascending = False)
```
